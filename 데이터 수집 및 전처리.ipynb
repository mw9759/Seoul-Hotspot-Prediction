### 환산임대료 데이터셋 전처리

````python
import pandas as pd
import numpy as np
import math
import re
````

````python
rt = pd.read_csv('환산임대료_final.csv', encoding='ANSI')
rt.drop('Unnamed: 0', axis=1, inplace=True)
````

````python
rt.head(10)
````

#### 결측값 0으로 대체

````python
missed_19 = [5, 14, 42, 114, 150, 158, 175, 347, 371, 405, 422]
for i in missed_19:
    rt.loc[i, '19년도'] = '0'
    
missed_20 = [114, 371, 422]
for i in missed_20:
    rt.loc[i, '20년도'] = '0'
    
missed_21 = [405, 422]
for i in missed_21:
    rt.loc[i, '21년도'] = '0'
````

#### 연도별 임대료 데이터 정수 타입으로 변형

````python
for i in range(len(rt)):
    rt['19년도'][i] = re.sub('\,', '', rt['19년도'][i])
    
for i in range(len(rt)):
    rt['20년도'][i] = re.sub('\,', '', rt['20년도'][i])
    
for i in range(len(rt)):
    rt['21년도'][i] = re.sub('\,', '', rt['21년도'][i])
    
````

````python
rt['19년도'] = rt['19년도'].astype(float)
rt['20년도'] = rt['20년도'].astype(float)
rt['21년도'] = rt['21년도'].astype(float)

rt['19년도'].replace(0, np.nan, inplace=True)
rt['20년도'].replace(0, np.nan, inplace=True)
rt['21년도'].replace(0, np.nan, inplace=True)

rt.head(20)
````

````python
print(rt.dtypes)
print(rt.isnull().sum())
print(rt.mean().round())
````

#### 결측치 평균으로 대체

````python
rt = rt.fillna(rt.mean().round())
rt.head(10)
````

#### 파생변수 생성_ 3년치 임대료 증감율 기하평균

````python
a = (rt['20년도'] / rt['19년도'])
b = (rt['21년도'] / rt['20년도'])
dv = pd.DataFrame({'20년도 증감율':a, '21년도 증감율':b})
````

````python
dv['증감율 기하평균'] = ((dv['20년도 증감율']*dv['21년도 증감율'])**(1/2) - 1) * 100
dv.head()
````

#### 환산임대료 데이터셋과 파생변수 결합

````python
fn = pd.concat([rt,dv], axis=1)
fn.head()
````

#### 컬럼명 변경 & 불필요한 컬럼 삭제

````python
fn.rename(columns={'21년도':'환산임대료(원, 21년도)'}, inplace=True)
fn.rename(columns={'증감율 기하평균':'증감율(%, 19-21년도)'}, inplace=True)
````

````python
fn.head()
````

````python
fn = fn.loc[:,['행정동_코드', '행정동', '환산임대료(원, 21년도)', '증감율(%, 19-21년도)']]
fn.head()
````

#### 증감율 컬럼 값 반올림하기

````python
fn['증감율(%, 19-21년도)'] = fn['증감율(%, 19-21년도)'].round(3)
````

#### 환산임대료(원, 21년도) 컬럼 타입 변경

````python
fn['환산임대료(원, 21년도)'] = fn['환산임대료(원, 21년도)'].astype('int64')
````

````python
fn.head()
````

### csv파일로 저장

````python
fn.to_csv('./환산임대료_preprocessed.csv')
````

# 문화공간(공연장추가 파일) 전처리

````python
#필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import requests, json
pd.set_option('display.max_rows', None)
#문화공간 정보 데이터 불러오기
df_cul = pd.read_csv("서울시 문화공간 정보.csv", encoding="cp949")
df_cul = df[["주제분류", "문화시설명","주소","위도","경도"]]
#문화공간 정보 데이터 결측치 확인
print(df_cul.isnull().sum(),"\n")
print(df_cul["위도"][df_cul["위도"].isnull()]) #df.loc[509]는 동작구청으로 삭제해도 될 것 같음
df_cul.loc[509]는 동작구청이므로 삭제
df_cul.drop([509], axis=0, inplace=True)
df_cul.reset_index(drop=True, inplace=True)
````

````python
#위도와 경도 받아서 주소 불러오는 함수
def get_address(lat, lng):
    url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x="+lng+"&y="+lat
    headers = {"Authorization": "KakaoAK c646e246bf7b31cec27b82f03e306284"}
    api_json = requests.get(url, headers=headers)
    address = json.loads(api_json.text)

    return address
````

````python
lat = df_cul["위도"].apply(str)
long = df_cul["경도"].apply(str)

ad = []
dong = []
code = []

for i, j in zip(lat,long):
    total = get_address(i,j)
    add = list(m["address_name"] for m in total["documents"])
    cod = list(m["code"] for m in total["documents"])
    
    don = list(m["region_3depth_name"] for m in total["documents"])
    ad.append(add[1])
    code.append(cod[1][0:8]) #행정동코드 10자리->8자리로 맞춤
    dong.append(don[1])

df_result = pd.DataFrame(data= list(zip(ad,dong,code)), columns=["지번주소","행정동","행정동코드"])

#받아온 지번주소, 행정동, 행정동코드를 기존 데이터프레임과 결합
df_merge = pd.merge(df_cul, df_result,left_index=True, right_index=True)
df_merge = df_merge[["주제분류","문화시설명","지번주소","행정동","행정동코드"]]
df_merge.head()
````

````python
#61인덱스~750인덱스는 서울 아니므로 행제거
df_merge = df_merge.drop([61,280,532,538,539,678,698,750], axis=0) 
df_1 = df_merge.reset_index(drop=False)
````

````python
#공연장 정보 데이터 불러오기
df_con = pd.read_csv("서울시 공연장 인허가 정보.csv", encoding="cp949")
df_con = df_con[["상세영업상태명","사업장명","지번주소","좌표정보(X)","좌표정보(Y)"]]
con = df_con.rename(columns = {"사업장명":"문화시설명"})
con = con[con["상세영업상태명"]== "영업중"]
con.isnull().sum() #결측치 확인 => 결측치 없는 지번주소로 행정동 불러오자
con.reset_index(drop=True, inplace=True)
````

````python
# 카카오API를 사용하여 주소->행정동 가져오기 (좌표도 가능)
def get_coord(address):
  try:
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK c646e246bf7b31cec27b82f03e306284"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    add = list(m["address"] for m in api_json["documents"])
    add1 = add[0]
    h_code = add1["h_code"][0:8] #행정동코드 10자리->8자리로 맞추기
    h_dong = add1["region_3depth_h_name"]
    return h_code, h_dong

  except IndexError:
    return np.nan, np.nan
````

````python
address = con["지번주소"]

h_code = []
h_dong = []

for i in address:
    coord = get_coord(i)
    h_code.append(coord[0])
    h_dong.append(coord[1])

con_hang = pd.DataFrame(data= list(zip(h_code, h_dong)), columns=["행정동코드","행정동"])

#받아온 행정동 데이터프레임과 기존 데이터프레임을 결합
con_merge = pd.merge(con, con_hang, left_index=True, right_index=True)
con_merge.head()
````
